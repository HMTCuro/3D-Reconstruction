```python
import cv2
import numpy as np
from functools import reduce

def process_name(index):
    if index <10:
        res="0000000"+str(index)
    elif index>=10 and index <100:
        res="000000"+str(index)
    elif index>=100 and index <1000:
        res="00000"+str(index)
    return res

colors = np.array([
    [128, 64, 192],  # tree
#     [64, 128, 128],  # car
#     [0, 192, 192],   # ac/box
    [128, 192, 192],   # ac/box
    [0, 128, 64],   # electric vehicle
    [64, 128, 0],   # trash can
])

for i in range(434):
    colored_img_dir = 'seg_out_dongbeiya_2/out_' + process_name(i) + ".png"
    colored_img = cv2.imread(colored_img_dir, 1)
    origin_img_dir = 'images/' + process_name(i) + ".jpg"
    origin_img = cv2.imread(origin_img_dir, 1)
    colored_img = cv2.cvtColor(colored_img, cv2.COLOR_BGR2RGB)
    origin_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
#     print(colored_img[:5, :5])
    for color in colors:
        colored_img_r = colored_img[:, :, 0]
        colored_img_g = colored_img[:, :, 1]
        colored_img_b = colored_img[:, :, 2]
        mask_r = np.where(colored_img_r == color[0], 1, 0)
        mask_g = np.where(colored_img_g == color[1], 1, 0)
        mask_b = np.where(colored_img_b == color[2], 1, 0)
        mask = mask_r * mask_g * mask_g
        indices = np.where(mask > 0)
#         print(indices)
        print(i, color, len(indices[0]))
#         if np.all(color == np.array([0, 64, 192])):  # box2
#             origin_img[indices] = [64, 0, 0]
#         elif np.all(color == np.array([64, 0, 128])):  # car2
#             origin_img[indices] = [64, 128, 128]
#         else:
        origin_img[indices] = color
    
    origin_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)
    cv2.imwrite('colored_dongbeiya/colored_' + process_name(i) + ".png", origin_img)
    
```
