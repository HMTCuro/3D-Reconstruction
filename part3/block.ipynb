```python
import cv2 as cv
import numpy as np
from copy import deepcopy

def process_name(index):
    if index < 10:
        res = "0000000" + str(index)
    elif index >= 10 and index < 100:
        res = "000000" + str(index)
    elif index >= 100 and index < 1000:
        res = "00000" + str(index)
    return res

for k in range(434):
    print(process_name(k))
    img_dir = "images/" + process_name(k) + ".jpg"
    source_img = cv.imread(img_dir)
    h, w = source_img.shape[:2]  # 3000, 4000
    h_block, w_block = 600, 800
    for i in range(h // h_block):
        for j in range(w // w_block):
#             print(i, j, i * w // w_block + j)
            block = deepcopy(source_img[i * h_block:i * h_block + h_block, j * w_block:j * w_block + w_block])
#     img_resize = cv.resize(source_img, (1200, 900))
            out_dir = "resized_images_dongbeiya_block/" + process_name(k) + "_" + str(i * w // w_block + j) + ".jpg"
            cv.imwrite(out_dir, block)
```
